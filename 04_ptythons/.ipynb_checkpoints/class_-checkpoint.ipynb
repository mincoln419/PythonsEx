```python
user_datas = [
    {"id" : "test", "pw" : "1234", "count" : 0},
    {"id" : "python", "pw" : "5678", "count" : 0}
]
```

```python
def need_login(func):
    def wrapper(*arg, **kwargs):
        #아이디 패스워드 확인
        id = input("아이디를 입력하세요: ")
        pw = input("패스워드를 입력하세요: ")
        flag = False
        for data in user_datas :
            if data["id"] == id :
                flag = True
                if data["pw"] == pw :
                    data["count"] += 1
                    print(func(*arg, **kwargs))
                    break;
                else : 
                    print("비밀번호가 일치하지 않습니다")
                    
        if flag == False:
            print("존재하지 않는 아이디 입니다")
    return wrapper
            
@need_login
def plus(num1, num2):
    return num1 + num2
```

```python
plus(3, 5)
```

```python
import random
# 스타크래프트 마린
class Marin :
    def __init__(self, hp = 40, ap = 5, name = "marin" + str(random.randint(1, 1000))) :
        self.hp = hp
        self.ap = ap
        self.name = name
    def attack(self, target):
        if(target.hp == 0):
            print("{name} 이(가) 이미사망하였습니다".format(name = target.name))
        else:
            minus = 0
            if target.hp >= self.ap :
                minus = self.ap
            else :
                minus = target.hp
            target.hp -= minus
            print("체력 {po} 이 깎였습니다. 현재 체력은 {now} 입니다".format(po = minus, now = target.hp))
            if target.hp == 0 :
                print("{name} 이(가) 사망하였습니다".format(name = target.name))
        

        
        
```

```python
marin1 = Marin(40, 5)
marin2 = Marin(40, 5)


```

```python
marin1.attack(marin2)
```

```python
print(marin1.hp, marin2.hp)
```

### 1. 상속
- 클래스의 기능을 가져다가 기능을 수정하거나 추가할때 사용하는 방법

```python
class Calculator:
    def __init__(self, num1, num2):
        self.num1 = num1
        self.num2 = num2
        
    def plus(self):
        return self.num1 + self.num2

    
```

```python
cal = Calculator(1, 2)
cal.plus()
```

```python
## 상속 : class(parent class)
class Calculator2(Calculator):
    def __init__(self, num1, num2):
        self.num1 = num1
        self.num2 = num2
    
    ###매서드오버라이드 - plus
    def plus(self):
        return self.num1**2 + self.num2**2
    
    def minus(self):
        return self.num1 - self.num2

    
```

```python
cal2 = Calculator2(2, 3)
cal2.plus(), cal2.minus()
```

### 2. super
- 부모 클래스에서 사용된 함수의 코드를 가져다가 자식 클래스의 함수에서 재사용할때..

```python
import random
# 스타크래프트 마린
class Marin :
    def __init__(self, hp = 40, ap = 5, name = "marin" + str(random.randint(1, 1000))) :
        self.hp = hp
        self.ap = ap
        self.name = name
    def attack(self, target):
        if(target.hp == 0):
            print("{name} 이(가) 이미사망하였습니다".format(name = target.name))
        else:
            minus = 0
            if target.hp >= self.ap :
                minus = self.ap
            else :
                minus = target.hp
            target.hp -= minus
            print("체력 {po} 이 깎였습니다. 현재 체력은 {now} 입니다".format(po = minus, now = target.hp))
            if target.hp == 0 :
                print("{name} 이(가) 사망하였습니다".format(name = target.name))
        

        
```

```python
class Marin2(Marin, Medic):
    
    def __init__(self):
        super(Marin, self).__init__() ##2개 이상있을때는 이렇게 쓰면됨
        self.max_health = 40
```

```python
marin2 = Marin2()
```

```python
print(marin2.max_health)
print(marin2.hp)
```

```python
class Marin3(Marin):
    
    def __init__(self):
        super().__init__() ##상속받는게 1개 있을 경우에는 이렇게 사용
        self.max_health = 40
```

### 3. class의 getter setter
- 객체 내부 변수에 접근할 때 특정 로직을 거쳐서 접근하는 방법

```python
class User:
    def __init__(self, first_name, last_name):
        self.first_name = first_name
        self.last_name = last_name
    def disp(self):
        print(self.first_name, self.last_name)
```

```python
user1 = User("andy", "kim")
```

```python
user1.disp()
```

```python
user1.first_name ="jhone"
```

```python
user1.disp()
```

```python
user1.first_name = 1
```

```python
user1.disp()
```

```python
## getter setter 추가

class User:
    def __init__(self, first_name, last_name):
        self.first_name = first_name
        self.last_name = last_name
    
    def setter(self, first_name):
        print("setter")
        if type(first_name) is str:
            ## isinstance(first_name, str)
            self.first_name = first_name
        else :
            print("error: not string type")
        
    def getter(self):
        print("getter")
        return self.first_name
    
    def disp(self):
        print(self.first_name, self.last_name)
        
    name = property(getter, setter)
    
```

```python
user1 = User("andy", "kim")
```

```python
user1.name
```

```python
user1.name = 1
```

```python
user1.name
```

```python
user1.name = "jhon"
```

```python
user1.name
```

### 4. non public
-- mangling 이라는 방법으로 다이렉트로 객체 변수에 접근하지 못하게 하는 방법

```python
class Calculator:
    def __init__(self, num1, num2):
        self.num1 = num1
        self.__num2 = num2 ## mangling
        
    def plus(self):
        return self.num1 + self.num2
    
    def getter(self):
        return self.num2
    
    # num2에 0이 들어가지 않도록함
    def setter(self):
        if(num2 == 0):
            num2 = 1
        self.num2 = num2
    
    def div(self):
        return self.num1 / self.num2
    
    number2 = property(getter, setter)
```

```python
cal = Calculator(1,2)
```

```python
cal.div()
```

```python
cal.num2
```

```python
cal.number2 = 0
```

```python
cal.num2
```

```python
cal.div()
```
