### Pandas Pivot
- 데이터 프레임의 컬럼 데이터에서 index, columnb, value를 선택해서 데이터 프레임을 만드는 방법
- `df.pivot(index, columns, values)` - group by 후 피벗
- `df.pivot_table(values, index, columns, aggfunc)`

### kaggle
- 데이터 분석, 모델을 경쟁할  수 있도록 만든 서비스


```python
### 1. 성별, 좌석등급에 따른 데이터의 수
df1 = titanic.groupby(["Sex", "Pclass"]).size().reset_index(name="counts")
df1
```

```python
result = df1.pivot("Sex", "Pclass", "counts")
result
```

```python
# pivot table 이용
titanic["counts"] = 1
titanic.tail(2)
```

```python
result = titanic.pivot_table("counts",["Pclass"], ["Survived"], aggfunc=np.sum)
```

```python
titanic.pivot_table("counts",["Sex"], ["Survived"], aggfunc=np.sum)
```

```python
# col 합계
result["total"] =result[0] + result[1]
```

```python
result
```

```python
# row 합계
result.loc["total"] = result.loc[1] + result.loc[2] + result.loc[3]
result
```
