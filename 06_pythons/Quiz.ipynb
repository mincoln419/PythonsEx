```python
import numpy as np
import pandas as pd
from makedata import *

```

```python
# make 모듈을 이용해서 데이터 프레임 만들기
# 딕셔너리의 리스트
datas = {}
datas["UserID"] = list(range(1, 9))
datas["Age"] = [get_age() for _ in range(8)]

names = []
while True :
    name = get_name()
    if name not in names :
        names.append(name)
    if len(names) >= 8:
        break
datas["Name"] = names
    

user_df = pd.DataFrame(datas)
```

```python
user_df2 = pd.DataFrame(columns = ["UserID", "Name", "Age"])

for idx in range(1, 9):
    name = get_name()
    while name in list(user_df2["Name"]):
        name = get_name()
    
    data = {"Name":name, "UserID": idx, "Age": get_age()}
    print(data)
    user_df2.loc[len(user_df2)] = data

user_df2
```

### quiz2 - money df

- 15개의 데이터
    - ID : 1 ~ 8 랜덤데이터
    - Money : 1000원 단위, 1000 ~ 20000까지의 숫자


```python
datas = {}
datas["ID"] = []
while True:
    id = np.random.randint(1, 9)
   # if id in datas["ID"] and len(datas["ID"]) < 8:
    #    continue
    datas["ID"].append(id)
    if len(datas["ID"]) == 15 :
        break;

datas["Money"] = np.random.randint(1, 20, 15) * 1000
money_df = pd.DataFrame(datas)
money_df
```

```python
ids = money_df["ID"].unique()
ids.sort()
ids
```

```python
user_df.rename(columns = {"UserID" : "ID"}, inplace = True)


#merge
result_df = user_df.merge(money_df)
result_df
```

```python
# groupby : sum ,size, min... 결과 series
money_list = result_df.groupby("Name").sum()["Money"].reset_index(name = "Money_Sum")
money_list
```

```python
# agg() ... 결과 Dataframe
money_list = result_df.groupby("Name").agg("sum").reset_index()[["Name", "Money"]]
money_list
```

```python
result_df = pd.merge(user_df, money_df, how ="outer")
result_df.fillna(value = 0, inplace = True)
result_df
```
